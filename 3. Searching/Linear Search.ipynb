```python
array = [10, 20, 80, 30, 60, 50, 110, 100, 130, 170]
# you need to search if x = 110 is present or not, if yes then return the index of the element
x = 110 # x is the required element
```

In this type of searching algo, we iterate over the entire array/list, and check whether the required
element is present or not.

```python
for i in range(len(array)):
    # len(array) --> 10 -- as there are 10 elements present in the array
    # this loop will run till 9 iterations, i.e it will run from 0 - 9 times, as in case of for 
    # loop the last point of the range is not considered. So basically we running 9 times.
    
    if array[i] == x:
        # Iterating over the entire array, and checking if the required element is present or not
        print('Index of the element is: ',i)
```

### Time Complexity:

**Worst case**: for Linear Search will be --> when we have to iterate the entire array/list.
So, if the size of the array is n, and the element to be find is at the last position of the list/array, OR the element is not in the list/array, then in such case, Linear Search will take 
**O(n)** time.

**Best Case:** For Linear Search the best case will be --> when the element to be find is at the first position of the array/list, in such case, we donot have to iterate the entire list/array as we will get the element in the first iteration itself.
Therefore in such case, time taken will be of order **O(1)**.

#### We can decrease the time complexity of Linear Search:

This can be done, by creating two pointers which point the first and the last position of the array/list, and then search from both the ends.

```python
pos = 0
right = len(array) - 1
for i in range(len(array)):
    if array[i] == x:
        print(i)
        break
    if array[right] == x:
        print(right)
        break
    right -= 1
    
    
```

By doing so, if we want the element is present at the last position, still it will take time of **O(1)**, but if the element is not found, then it will take the time of **O(n/2)** == **O(n)**
